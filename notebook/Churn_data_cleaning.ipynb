```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('D:/churn/data.csv')

# Show first 5 rows of data (sample data)
print("Sample data:")
print(df.head())

# Check data shape and columns info
print("\nData shape (rows, columns):", df.shape)
print("\nColumns and types:")
print(df.info())

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# View descriptive statistics (numeric columns)
print("\nDescriptive statistics:")
print(df.describe())

# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"\nDuplicate rows count: {duplicates}")

# If duplicates exist, you can drop them:
if duplicates > 0:
    df = df.drop_duplicates()
    print("Duplicates removed.")

print(df.isnull().sum())
```

```python
df = pd.read_csv('D:/churn/data.csv')

# Replace spaces in column names with underscore
df.columns = df.columns.str.replace(' ', '_')

print(df.info())
```

```python
# Convert to numeric
df['Total_Charges'] = pd.to_numeric(df['Total_Charges'], errors='coerce')

# Calculate mean
mean_val = df['Total_Charges'].mean()

# Fill NaN without inplace to avoid warning
df['Total_Charges'] = df['Total_Charges'].fillna(mean_val)

# Churn_Reason की null values को "No Information" से replace किया
df['Churn_Reason'] = df['Churn_Reason'].fillna("No Information")
```
